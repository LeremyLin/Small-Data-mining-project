```
import random 
# Read the data file, split it to train and test set. Get the result from those two set
def readFile(file):
  f = open(file)
  lines = f.readlines()
  numOfData = len(lines)
  numTest = int(numOfData*0.1)
  numTrain = numOfData-numTest
  random.shuffle(lines)
  lines[:] = [line.rstrip('\n') for line in lines]
  testLine = lines[:numTest+1]
  trainLine = lines[numTest+1:]
  testArr = [[0 for i in range(7)] for j in range(numTest+1)]
  trainArr = [[0 for i in range(7)] for j in range(numTrain+1)]
  # Get two dimension for test set
  for i in range(len(testLine)):
    l = testLine[i].split(',')
    for k in range(7):
      testArr[i][k] = int(l[k])
  # Get two dimension for train set
  for i in range(len(trainLine)):
    l = trainLine[i].split(',')
    for k in range(7):
      trainArr[i][k] = int(l[k])
  return testArr,trainArr,numTest,numTrain
```

```
# Get train and test set
testArr,trainArr,numTest,numTrain = readFile('car.txt')
```

```
import numpy as np
# Normalize the data
def normalization(trainArr):
  maxVal = [4,4,5,6,3,3]
  minVal = [1,1,2,2,1,1]
  newTrainArr = trainArr
  for i in range(len(trainArr)):
    for j in range(6):
       newTrainArr[i][j] = float(trainArr[i][j] - minVal[j])/float(maxVal[j]-minVal[j])
  return newTrainArr
```

```
# Normalization training set
newTrainArr = normalization(trainArr)
```

```
## KNN method
def knnCal(inputList,newTrainArr,numK):
  maxVal = [4,4,5,6,3,3]
  minVal = [1,1,2,2,1,1]
  newInputList = inputList
  resultDict = {}
  # Convert test data to normalized form
  for j in range(6):
       newInputList[j] = float(inputList[j] - minVal[j])/float(maxVal[j]-minVal[j])

  distanceStore = []
  resultStore = []
  maxDis = 10000
  for i in range(len(newTrainArr)):
    distance = 0
    # Get the sum of distance for each attribute between test and trian data
    for j in range(6):
      distance+=(float(float(newTrainArr[i][j]-newInputList[j])**2))
    # Decide how to store the result
    if(distance<maxDis):
      # Still get space to store the value
      if(len(distanceStore)<numK):
        distanceStore.append(distance)
        resultStore.append(newTrainArr[i][-1])
      else:
        maxIndex = distanceStore.index(max(distanceStore))
        distanceStore[maxIndex] = distance
        resultStore[maxIndex] = newTrainArr[i][-1]
  
  predictResult = max(set(resultStore), key=resultStore.count)
  return predictResult
```

```
# Below is to do the prediciton on test set to test the model
count = 0
# Store all predcited result
rightStoreCount = []
# Store correct result
allStoreCount = []
# Test for all test set
for i in range(len(testArr)):
  predictResult = knnCal(testArr[i],newTrainArr,10)
  allStoreCount.append(predictResult)
  if(predictResult==testArr[i][6]):
    count+=1
    rightStoreCount.append(predictResult)

# Print accuracy
print(count/len(testArr))
```
