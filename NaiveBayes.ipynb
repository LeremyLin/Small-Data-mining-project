```
import random 
# Read the data file, split it to train and test set. Get the result from those two set
def readFile(file):
  f = open(file)
  lines = f.readlines()
  numOfData = len(lines)
  numTest = int(numOfData*0.1)
  numTrain = numOfData-numTest
  random.shuffle(lines)
  lines[:] = [line.rstrip('\n') for line in lines]
  testLine = lines[:numTest+1]
  trainLine = lines[numTest+1:]
  testArr = [[0 for i in range(7)] for j in range(numTest+1)]
  trainArr = [[0 for i in range(7)] for j in range(numTrain+1)]
  # Get two dimension for test set
  for i in range(len(testLine)):
    l = testLine[i].split(',')
    for k in range(7):
      testArr[i][k] = int(l[k])
  # Get two dimension for train set
  for i in range(len(trainLine)):
    l = trainLine[i].split(',')
    for k in range(7):
      trainArr[i][k] = int(l[k])
  return testArr,trainArr,numTest,numTrain
```

```
testArr,trainArr,numTest,numTrain = readFile('car.txt')
```

```
# Calculate four probability for input test data
def calFourPro(listInput,trainArr):
  # Store the matching number from four different result
  store1 = [0 for i in range(6)]
  store2 = [0 for i in range(6)]
  store3 = [0 for i in range(6)]
  store4 = [0 for i in range(6)]
  result = [0 for i in range(4)]
  # Each line in training set
  for case in trainArr:
    if case[6]==1:
      result[0]+=1
      for i in range(len(case)-1):
        if(case[i]==listInput[i]):
         store1[i]+=1   
    elif case[6]==2:
      result[1]+=1
      for i in range(len(case)-1):
        if(case[i]==listInput[i]):
         store2[i]+=1 
    elif case[6]==3:
      result[2]+=1
      for i in range(len(case)-1):
        if(case[i]==listInput[i]):
         store3[i]+=1 
    else:    
      result[3]+=1
      for i in range(len(case)-1):
        if(case[i]==listInput[i]):
         store4[i]+=1 

  # Get the probability from each element from four result 
  storeProb1 = [i/len(trainArr) for i in store1]
  storeProb2 = [i/len(trainArr) for i in store2]
  storeProb3 = [i/len(trainArr) for i in store3]
  storeProb4 = [i/len(trainArr) for i in store4]

  
  # Probability for last acceptance in result
  for i in range(4):
    result[i]/=len(trainArr)

  # Probability from each result
  prob=[1 for i in range(4)]

  # Multiply into the previous reuslt
  for i in range(4):
    prob[i]*=result[i]

  for i in range(6):
    prob[0]*=storeProb1[i]  
    prob[1]*=storeProb2[i]
    prob[2]*=storeProb3[i]
    prob[3]*=storeProb4[i]
  

  max = 0
  index = -1
  for i in range(4):
    if(prob[i]>max):
      max=prob[i]
      index = i+1
  return index
  
```

```
count = 0
storeCount = []
for i in range(len(testArr)):
  index = calFourPro(testArr[i],trainArr)
  if(index==testArr[i][6]):
    count+=1
    storeCount.append(index)

print(count/len(testArr))
```
